多个地区循环遍历转换

```python
import os
import pandas as pd
import json

ori_folder = './ori_data'
dst_folder = './processed_data'

# 确保目标文件夹存在
os.makedirs(dst_folder, exist_ok=True)

# 遍历所有csv文件
for filename in os.listdir(ori_folder):
    if filename.endswith('.csv'):
        file_path = os.path.join(ori_folder, filename)
        try:
            data = pd.read_csv(file_path)
            reshaped_data = data.transpose()
            reshaped_data.columns = reshaped_data.iloc[0]
            reshaped_data = reshaped_data.drop(reshaped_data.index[0])

            # 有的csv文件格式可能不同，这里以你的处理方式为主
            # 提取daily字段，转换成DataFrame
            daily_data = json.loads(reshaped_data.iloc[0]['daily'])
            daily_df = pd.DataFrame(daily_data)
            daily_df['time'] = pd.to_datetime(daily_df['time'])

            # 保存
            out_path = os.path.join(dst_folder, filename)
            daily_df.to_csv(out_path, index=False)
            print(f"Processed and saved: {filename}")

        except Exception as e:
            print(f"Failed to process {filename}: {e}")
```

单个地区数据转换

```python
import pandas as pd
import json

# Load the CSV file to inspect its structure
file_path = './ori_data/xinjiang.csv'
data = pd.read_csv(file_path)

# Reshape the data so that features (e.g., latitude, longitude) become columns
reshaped_data = data.transpose()  # Transpose the dataset

# Set the first row as the header (feature names) and drop the previous header
reshaped_data.columns = reshaped_data.iloc[0]
reshaped_data = reshaped_data.drop(reshaped_data.index[0])

# Extract the "hourly" and "daily" columns, which contain JSON-like strings
hourly_data = json.loads(reshaped_data.iloc[0]['hourly'])  # Convert the "hourly" string to a JSON object
daily_data = json.loads(reshaped_data.iloc[0]['daily'])    # Convert the "daily" string to a JSON object

# Create a DataFrame for the 'daily' data, extracting the relevant information
daily_df = pd.DataFrame(daily_data)

# Convert the 'time' column to datetime
daily_df['time'] = pd.to_datetime(daily_df['time'])

# Display the processed daily DataFrame
daily_df.head()

# Optionally save the processed daily data to a CSV file
daily_df.to_csv('./processed_data/xinjiang.csv', index=False)
```
